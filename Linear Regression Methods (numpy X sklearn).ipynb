# Linear Regression using numpy

<img src='images/linreg1.png'>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#area
x = [1.2,2.6,3.2,1.8,2.5,3.6,4.2,6.8,6.4,7.2,8.2,9.1,8.2,4.2,5.6,3.9,1.5]
#price
y = [180,290,360,170,230,320,390,580,650,690,790,850,860,470,620,420,190]

x = np.array(x)
y = np.array(y)
```

```python
#random initialization of m and c
m = 10
c = 20
learning_rate = 0.03

for i in range(500):
    yhat = m*x+c
    E = np.mean(np.square(y-yhat))
    m = m-learning_rate*2*np.mean(x*(yhat-y))
    c = c-learning_rate*2*np.mean(yhat-y)
    print(E)

#objective: find error (and m&c) by using iteration
```

```python
print(m)
print(c)
```

```python
def ypred(x):
    yhat = m*x+c
    return yhat
```

```python
plt.scatter(x,y)
plt.plot(x,ypred(x),c='r')
plt.show()
```

# Linear Regression using sklearn

```python
#area
x = [1.2,2.6,3.2,1.8,2.5,3.6,4.2,6.8,6.4,7.2,8.2,9.1,8.2,4.2,5.6,3.9,1.5]
#price
y = [180,290,360,170,230,320,390,580,650,690,790,850,860,470,620,420,190]

x = np.array(x).reshape(17,1)
y = np.array(y)
#sklearn expect features(x) to be in 2D format (rows and columns)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
print('m = ',model.coef_)
print('c = ',model.intercept_)
```

```python
ypred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
plt.show()
```

# Gradient Descent Approach for linear regression

```python
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(learning_rate='invscaling',max_iter=5000)
model.fit(x,y)
```

```python
print('m: ',model.coef_)
print('c: ',model.intercept_)
```

```python
ypred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
