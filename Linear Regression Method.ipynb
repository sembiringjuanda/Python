# Linear Regression using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
#area
x = [1.2,2.6,3.2,1.8,2.5,3.6,4.2,6.8,6.4,7.2,8.2,9.1,8.2,4.2,5.6,3.9,1.5]
#price
y = [180,290,360,170,230,320,390,580,650,690,790,850,860,470,620,420,190]

x = np.array(x)
y = np.array(y)
```

```python
m = 10
c = 20
lr = 0.03

for i in range(500):
    yhat = m*x+c
    E = np.mean(np.square(y-yhat))
    m = m - lr*2*np.mean((yhat-y)*x)
    c = c - lr*2*np.mean(yhat-y)
    print(E)
```

```python
print(m)
print(c)
```

```python
def predict(x):
    yhat = m*x+c
    return yhat
```

```python
ypredict = predict(x)
plt.scatter(x,y)
plt.plot(x,ypredict,c='r')
plt.show()
```

# Linear Regression using sklearn

```python
#x needs to be reshaped to array because machine can only recognize it as array
x = np.array(x).reshape(17,1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
#m value
model.coef_
```

```python
#c value
model.intercept_
```

```python
ypred = model.predict(x)
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
plt.show()
```

# Linear regression using gradient descent

```python
from sklearn.linear_model import SGDRegressor
model2 = SGDRegressor(learning_rate='invscaling', max_iter=5000)
model2.fit(x,y)
```

```python
print(model2.coef_)
print(model2.intercept_)
```
